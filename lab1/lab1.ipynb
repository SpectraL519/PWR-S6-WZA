### Exercise 3

```python
import matplotlib.pyplot as plt
import numpy as np
from typing import Iterable


Iterable2D = Iterable[Iterable]

def plot_3d_algebraic_variety(
    X: Iterable2D, Y: Iterable2D, Z: Iterable2D,
    title: str = ""
):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.set_title(title)

    for (x, y, z) in zip(X, Y, Z):
        ax.plot_surface(x, y, z, cmap='viridis')

    return ax


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
z = np.linspace(-5, 5, 100)
x_mesh, y_mesh = np.meshgrid(x, y)

# V(z - x^2 - y^2)
# Paraboida obrotowa
z_mesh = x_mesh ** 2 + y_mesh ** 2
ax = plot_3d_algebraic_variety(X=[x_mesh], Y=[y_mesh], Z=[z_mesh], title="V(z - x^2 - y^2)")
plt.show();

# V(z^2 - x^2 - y^2)
z_mesh = np.sqrt(x_mesh ** 2 + y_mesh ** 2)
ax = plot_3d_algebraic_variety(
    X=[x_mesh] * 2,
    Y=[y_mesh] * 2,
    Z=[z_mesh, -z_mesh],
    title="V(z^2 - x^2 - y^2)"
)
plt.show();

# V(z - x^2 + y^2)
z_mesh = x_mesh ** 2 - y_mesh ** 2
ax = plot_3d_algebraic_variety(X=[x_mesh], Y=[y_mesh], Z=[z_mesh], title="V(z - x^2 + y^2)")
plt.show();

# V(xz, yz)
zero_plane = np.zeros_like(x_mesh)
ax = plot_3d_algebraic_variety(
    X=[x_mesh],
    Y=[y_mesh],
    Z=[zero_plane],
    title="V(xz, yz)"
)
ax.plot(
    [0, 0],  # x values
    [0, 0],  # y values
    [-5, 5], # z values
    linewidth=5, color='lime'
)
plt.show();
```

### Exercise 4

```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2 * np.pi, 1000)

radius = np.sin(2 * theta)

x = radius * np.cos(theta)
y = radius * np.sin(theta)

plt.figure(figsize=(8, 8))
plt.plot(x, y, color='blue')
plt.axis('equal')
plt.grid(True)
plt.show();
```

### Konwersja współrzędnych biegunowych na kartezjańskie:

Z wzoru $r(\theta) = \sin(2\theta) \implies r(\theta) = 2\sin(\theta)\cos(\theta)$

Możemy zatem wyznaczyć $r^3(\theta) = r^2(\theta) \cdot 2\sin(\theta)\cos(\theta) = 2(r(\theta)\sin(\theta))(r(\theta)\cos(\theta))$

Wiemy także, że $x(\theta) = r \cdot \cos(\theta) \land y(\theta) = r \cdot \sin(\theta)$. 

Stąd mamy: $r^3(\theta) = 2xy$

Ale wiemy, że $r(\theta) = \sqrt{x^2 + y^2} \implies r^3(\theta) = (x^2 + y^2)^\frac{3}{2}$

Zatem $(x^2 + y^2)^\frac{3}{2} = 2xy \implies (x^2 + y^2)^3 = 4x^2y^2$

```python
x = np.linspace(-1.1, 1.1, 500)
y = np.linspace(-1.1, 1.1, 500)

X, Y = np.meshgrid(x, y)
Z = (X ** 2 + Y ** 2) ** 3 - 4 * (X ** 2 * Y ** 2)

plt.figure(figsize=(5, 5))
plt.contour(X, Y, Z, levels=[0])
plt.title("$(x^2 + y^2)^3 = 4x^2y^2$")
plt.show();
```
